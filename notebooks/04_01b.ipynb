```python
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import DataFrame
```

```python
data = {'names': ['steve', 'john', 'richard', 'sarah', 'randy', 'micheal', 'julie'],
        'age':[20, 22, 20, 21, 24, 23, 22],
        'gender':['Male', 'Male', 'Male', 'Female', 'Male', 'Male', 'Female'],
        'rank':[2, 1, 4, 5, 3, 7, 6]}
```

```python
df = DataFrame(data)
```

```python
df
```

### Matplotlib's Bar Chart

```python
plt.bar(df['names'],df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing Ages')
```

```python
seaborn_plot = sns.barplot(data = df, x = 'names', y='age')
seaborn_plot.set_title('Comparing Ages')
plt.show()
```

### Line Plot Matplotlib

```python
plt.plot(df['names'],df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title("Comparing Ages")
plt.show()
```

### Line Plot Seaborn

```python
seaborn_plot = sns.lineplot(data=df, x='names', y='age')
seaborn_plot.set_title("Comparing Ages")
plt.show()
```

### Pie Chart Matplotlib

```python
plt.pie(df['age'],labels=df['names'])
plt.title('Age Comparison')
plt.show()
```

### Pie Chart Seaborn

```python
colors = sns.color_palette('pastel')[0:5]
plt.pie(df['age'],labels=df['names'],colors=colors)
plt.show()
```
