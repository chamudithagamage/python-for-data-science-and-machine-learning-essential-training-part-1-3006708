```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

### Concatenating data

```python
DF = DataFrame(np.arange(36).reshape(6,6))
```

```python
DF
```

```python
DF_2 = DataFrame(np.arange(15).reshape(5,3))
DF_2
```

```python
DF_3 = pd.concat([DF,DF_2],axis=1)
```

```python
DF_3
```

```python
pd.concat([DF,DF_2],axis=0)
```

#### Adding data

```python
series_obj = Series(np.arange(6))
series_obj.name = "added_variable"
series_obj
```

```python
variable_added = DataFrame.join(DF, series_obj)
```

```python
variable_added
```

```python
added_variable = pd.concat([variable_added,variable_added], ignore_index = False)
```
